## Question 1

**Step 1. Import the necessary libraries**

```python
import pandas as pd
```

**Step 2. Import the dataset from this address.**

**Step 3. Assign it to a variable called users**

```python
# the below mentioned is the link from where the dataset is taken.

# Assigned data to the variabled called users.

users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

**Step 4. Discover what is the mean age per occupation**


```python
# I have grouped it by occupation and then used the column name to call it

mean_age_per_occupation = users[['occupation', 'age']].groupby('occupation').mean()

# the mean_age_per _occupation will disply the result of mean age per occupation

mean_age_per_occupation
```

**Step 5. Discover the Male ratio per occupation and sort it from the most to the least**

```python
#Extracting the columns of gender and occupation to be stored in dataset
dataset = users[['gender', 'occupation']]

# finding the data of gender which are male
malegender = dataset.loc[dataset['gender'] == 'M']

#Group by used for occupation
grouped_occupation_data = dataset.groupby('occupation').count()

# Group by used for gender which are male
grouped_male_data = malegender.groupby('occupation').count()

# Calculating male ratio
male_ratio = grouped_male_data / grouped_occupation_data

#Sorting done on the basis of gender and in descending value
male_ratio.sort_values(ascending=False, by='gender')
```

**Step 6. For each occupation, calculate the minimum and maximum ages**

```python
#Extracting the columns of gender and occupation to be stored in dataset
dataset = users[['occupation', 'age']]

# Grouping the occupation by minimum age
Minimum_age_of_occupation = dataset.groupby('occupation').min()

# Grouping the occupation by maximum age
Maximum_age_of_occupation = dataset.groupby('occupation').max()

#Combining both the results
answer = pd.concat([Maximum_age_of_occupation, Minimum_age_of_occupation], axis=1)


answer
```

**Step 7. For each combination of occupation and sex, calculate the mean age**

```python
#Extracting the columns of gender and occupation to be stored in dataset
dataset = users[['occupation', 'gender', 'age']]

#Calculating mean age for each occupation and gender combination
mean_age = dataset.groupby(['occupation', 'gender']).mean('age')

mean_age
```

**Step 8. For each occupation present the percentage of women and men**

```python
#Grouping by occupation and gender
total_percentage = dataset.groupby('occupation')['gender'].value_counts(normalize=True)*100

total_percentage
```

## Question 6

**Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!**

```python
import pandas as pd
import matplotlib.pyplot as plt

#Reading data from the given link
dataset = pd.read_csv('us-marriages-divorces-1867-2014.csv')

dataset 
```

```python
# No axes for this figure which is also empty
figure = plt.figure() 

#Setting size for the figure
plt.rcParams["figure.figsize"] = [15, 10] 

# Set x & y axis and legends
dataset.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color = 'r', fontsize = 24)

#Setting labels
plt.xlabel('Year', color = 'b', fontsize = 18)
plt.ylabel('Capita', color = 'b', fontsize = 18)

plt.show()
```
